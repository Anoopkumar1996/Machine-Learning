# Support Vector Machine

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
iris= load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df['target']=iris.target
```

```python
df.head()
```

```python
iris.target_names
```

```python
df[df.target==2].head()
```

```python
df.shape
```

```python
df['flower_name'] = df.target.apply(lambda x:iris.target_names[x])
df
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
df0
```

```python
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],marker='.')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],marker='*')
```

```python
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],marker='.')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],marker='*')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['target','flower_name'],axis='columns')
```

```python
y= df.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model= SVC(C=10)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
